```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find('div',class_="relative group z-10").findChildren()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(results)
```

```python
sk = df.to_csv(index=False)
```
